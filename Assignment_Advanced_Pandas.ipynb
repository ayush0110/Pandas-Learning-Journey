```python
# Business Insights

# East dominates high-value sales overall — East leads with 1,66,052 in total high-value Sales (Sales > 3000), well ahead of South (1,41,048), West (1,16,551), and North (98,062). This is driven overwhelmingly by Laptops (44,096) — the single biggest cell in the entire table.

# Laptops and Phones are the premium-ticket products — Phones bring in the highest overall total (88,041) across regions, followed closely by Keyboard (77,924) and Laptop (72,508). Since we filtered for Sales > 3000, this tells you these categories most frequently produce large individual transactions — likely bulk orders or premium-priced units.

# Regional product strengths differ sharply, suggesting localized demand:

# East is the clear Laptop hub (44,096 — more than double any other region's Laptop total)
# West stands out in Keyboards (33,380 — the highest single Keyboard figure by far)
# North leads in Mouse sales (22,209) despite being the weakest region overall
# South shows the most balanced spread across all products — no single category dominates

# Printers and Headphones are consistently the weakest high-value categories across nearly every region, suggesting these products rarely generate large single transactions (likely lower unit prices, so they don't clear the 3000 threshold often).

# Recommendation angle: East's inventory/marketing strategy for Laptops seems to be working — worth investigating what's different there (pricing, customer base, promotions) and considering replicating it in North, which is comparatively lagging in nearly every product category.
```
